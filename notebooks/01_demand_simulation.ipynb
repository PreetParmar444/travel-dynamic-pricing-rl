```python
prices = [80, 100, 120, 140, 160]

for p in prices:
    print(
        p,
        calculate_purchase_probability(
            p,
            days_left=30
        )
    )
```

```python
import matplotlib.pyplot as plt

prices = range(80, 161, 5)

probabilities = [
    calculate_purchase_probability(price, 15)
    for price in prices
]

plt.figure(figsize=(8,5))
plt.plot(prices, probabilities)
plt.xlabel("Price")
plt.ylabel("Purchase Probability")
plt.title("Effect of Price on Customer Demand")
plt.grid(True)
plt.show()
```

```python
days = range(30,0,-1)

probabilities = [
    calculate_purchase_probability(
        100,
        d
    )
    for d in days
]

plt.figure(figsize=(8,5))
plt.plot(days, probabilities)
plt.xlabel("Days Left")
plt.ylabel("Purchase Probability")
plt.title("Demand Increases Near Departure")
plt.gca().invert_xaxis()
plt.grid(True)
plt.show()
```
